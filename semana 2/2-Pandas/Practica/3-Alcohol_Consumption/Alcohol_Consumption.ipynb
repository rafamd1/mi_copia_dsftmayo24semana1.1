### Step 8. Print the mean, min and max values for spirit consumption per continent.

```python
drinks2.groupby('continent')[['beer_servings', 'wine_servings', 'spirit_servings',  'total_litres_of_pure_alcohol']].describe()
```

```python
drinks2.groupby('continent')[['beer_servings', 'wine_servings', 'spirit_servings',  'total_litres_of_pure_alcohol']].mean()
```

```python
drinks2.groupby('continent')[['beer_servings', 'wine_servings', 'spirit_servings',  'total_litres_of_pure_alcohol']].min()
```

```python
drinks2.groupby('continent')[['beer_servings', 'wine_servings', 'spirit_servings',  'total_litres_of_pure_alcohol']].max()
```

CORRECCION:

ESTA MAL, ME PEDIAN SPIRIT SERVINGS

```python
drinks.groupby('continent')['spirit_servings'].agg([min, max, np.mean])
```

```python
drinks.groupby('continent')['spirit_servings'].agg([min, max, np.mean]).sort_values("mean", ascending = False)
```

```python
drinks.groupby("continent").agg({"beer_servings": [np.mean, np.median],
                                 "spirit_servings": [np.min, np.max, np.mean]})
# Ordenar de forma descendente por la mediana de beer_servings
```

```python
drinks3= drinks.groupby("continent").agg({"beer_servings": [np.mean, np.median],
                                 "spirit_servings": [np.min, np.max, np.mean]})
# Ordenar de forma descendente por la mediana de beer_servings
```

```python
drinks3[("beer_servings", "median")]
```

```python
drinks3.sort_values(("beer_servings", "median"), ascending= False)
```
